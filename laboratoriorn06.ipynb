```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return 2 * x**2

# np.arange(start, stop, step) para darnos una línea más suave.
x = np.arange(0, 5, 0.001)
y = f(x)
plt.plot(x, y)

# El punto y el punto "suficientemente cercano"
p2_delta = 0.0001
x1 = 2
x2 = x1 + p2_delta
y1 = f(x1)
y2 = f(x2)
print((x1, y1), (x2, y2))

# Aproximación de la derivada y la intersección y para la línea tangente
approximate_derivative = (y2 - y1) / (x2 - x1)
b = y2 - approximate_derivative * x2

# Ponemos el cálculo de la línea tangente en una función para que podamos llamarla
# múltiples veces para diferentes valores de x
# approximate_derivative y b son constantes para una función dada
# por lo tanto, se calculan una vez antes de esta función
def tangent_line(x):
    return approximate_derivative * x + b

# Graficando la línea tangente
# +/- 0.9 para dibujar la línea tangente en nuestro gráfico
# luego calculamos el y para el x dado usando la función de la línea tangente
# Matplotlib dibujará una línea para nosotros a través de estos puntos
to_plot = [x1 - 0.9, x1, x1 + 0.9]
plt.plot(to_plot, [tangent_line(i) for i in to_plot])

print('Aproximación de la derivada para f(x)',
      f'donde x = {x1} es {approximate_derivative}')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return 2 * x**2

# np.arange(inicio, fin, paso) para darnos una curva más suave
x = np.array(np.arange(0, 5, 0.001))
y = f(x)
plt.plot(x, y)

colores = ['k', 'g', 'r', 'b', 'c']

def approximate_tangent_line(x, approximate_derivative):
    return (approximate_derivative * x) + b

for i in range(5):
    p2_delta = 0.0001
    x1 = i
    x2 = x1 + p2_delta
    y1 = f(x1)
    y2 = f(x2)
    
    print((x1, y1), (x2, y2))
    
    approximate_derivative = (y2 - y1) / (x2 - x1)
    b = y2 - (approximate_derivative * x2)
    
    to_plot = [x1 - 0.9, x1, x1 + 0.9]
    plt.scatter(x1, y1, c=colores[i])
    plt.plot([point for point in to_plot],
             [approximate_tangent_line(point, approximate_derivative)
              for point in to_plot],
             c=colores[i])
    
    print('Derivada aproximada para f(x)',
          f'donde x = {x1} es {approximate_derivative}')

plt.show()
```
