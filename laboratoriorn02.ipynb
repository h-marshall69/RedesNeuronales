```python
# Capa Densa o totalmente conectada
class Layer_Dense:
    # Inicialización de la capa
    def __init__(self, n_inputs, n_neurons):
        # Inicializar pesos y sesgos
        self.weights = 0.01 * np.random.randn(n_inputs, n_neurons)
        self.biases = np.zeros((1, n_neurons))

    # Paso hacia adelante
    def forward(self, inputs):
        # Calcular los valores de salida a partir de las entradas, pesos y sesgos
        self.output = np.dot(inputs, self.weights) + self.biases
```

```python
# Capa Densa o totalmente conectada
class Layer_Dense:
    # Inicialización de la capa
    def __init__(self, n_inputs, n_neurons):
        self.weights = 0.01 * np.random.randn(n_inputs, n_neurons)
        self.biases = np.zeros((1, n_neurons))
    # Paso hacia adelante
    def forward(self, inputs):
        self.output = np.dot(inputs, self.weights) + self.biases
```

```python
# Crear conjunto de datos
X, y = spiral_data(samples=100, classes=3)
# Crear una capa densa con 2 características de entrada y 3 valores  de salida
dense1 = Layer_Dense(2, 3)
# Realizar un pase hacia adelante de nuestros datos de  entrenamiento a través de esta capa
dense1.forward(X)
# Veamos la salida de las primeras muestras:
print(dense1.output[:5])
```
