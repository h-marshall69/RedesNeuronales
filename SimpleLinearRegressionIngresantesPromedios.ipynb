```python
#Librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#DATASET
data = '../input/d/japondo/simple-linear-regression/Simple linear regression.csv'
df = pd.read_csv(data)
```

```python
#5 Primeros datos
df.head
```

```python
plt.scatter(df['SAT'], df['GPA'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('DATOS DE EJEMPLOS PARA LA REGRESION')
plt.show()
```

```python
x = df['SAT']
y = df['GPA']
```

```python
m = 0.001 #Pendiente de la curva
b = 1.5 #Ordenada al origen
y_pred = m * x + b
```

```python
plt.scatter(x, y, color= 'blue', label='Datos')
plt.plot(x, y_pred, color='red', label='Linea de regresion')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regresion lineal  con linea definida manualmente')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.array(x).reshape(-1,1)
print(x)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=4)
X_train
```

```python
from sklearn.linear_model import LinearRegression

#Creamos el modelo
model = LinearRegression()

#Entrenar(Ajustar el modelo de datos)
model.fit(X_train, y_train)

#Realizar la prediccion
y_pred = model.predict(X_test)
```

```python
from sklearn import metrics
#Calculamos r2

print('r2', metrics.r2_score(y_test, y_pred))
```

```python
from sklearn.metrics import mean_squared_error

#RMSE TEST / ERRRO CUADRATICO MEDIO
RMSE = np.sqrt(mean_squared_error(y_pred, y_test))
RMSE
```

```python
plt.scatter(x, y, color= 'blue', label='Datos')
plt.plot(X_test, y_pred, color='red', label='Linea de regresion')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regresion lineal y su recta')
plt.legend()
plt.grid(True)
plt.show()
```
